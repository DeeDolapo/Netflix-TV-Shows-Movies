#### Research question 1
##### Which year is the highest no of tv shows released

```python
df_movies.dtypes['release_year']
```

```python
df_movies['release_year'].value_counts()
```

```python
plt.figure(figsize = (10,5), dpi = 150)
df_movies['release_year'].hist(bins = 20)
plt.title("No of movies released in a year", size = 10)
plt.xlabel('release_year')
plt.ylabel('Frequency')
plt.show
```

##### From the above, highest number of movies were released in year 2018

#### Research question 2
##### What is the name of the show with the highest runtime

```python
df_movies.head(5)
```

```python
Highest_runtime= df_movies.groupby(by = 'runtime').max()[['title']].sort_values(by = ['title'], ascending = True).reset_index()
fig = plt.figure(figsize = (16,9))
plt.title("Movie runtime", size = 25)
axis = sns.barplot(data = Highest_runtime.iloc[:15], y = 'title', x = 'runtime', linewidth = 2, edgecolor = 'green')
plt.xlabel('runtime')
plt.ylabel('title')
plt.show
```

##### From the above, jeen-yuhs is the tv show with the highest runtime

```python
df_movies[df_movies['title'].isin(['1994'])]
```

##### Research question 3
###### which movie has the highest season

```python
Highest_season= df_movies.groupby(by = 'seasons').max()[['title']].sort_values(by = ['title'], ascending = True).reset_index()
fig = plt.figure(figsize = (16,9))
plt.title("Movie seasons", size = 25)
axis = sns.barplot(data = Highest_season.iloc[:15], y = 'title', x = 'seasons', linewidth = 2, edgecolor = 'green')
plt.xlabel('seasons')
plt.ylabel('title')
plt.show
```

```python
df_movies[df_movies['title'].isin(['Wheel of Fortune'])]
```

```python
df_movies[df_movies['title'].isin(['Survivor'])]
```

##### From the above, Movie title 'Survivor' has the most season

##### Research question 6
###### What show has the highest imdb score

```python
Highest_imdb_score= df_movies.groupby(by = 'imdb_score').max()[['title']].sort_values(by = ['title'], ascending = True).reset_index()
fig = plt.figure(figsize = (16,9))
plt.title("Movie Imdb Score", size = 25)
axis = sns.barplot(data = Highest_imdb_score.iloc[:15], y = 'title', x = 'imdb_score', linewidth = 2, edgecolor = 'green')
plt.xlabel('imdb_score')
plt.ylabel('title')
plt.show
```

##### From the above, Breaking bad has the highest imdb score

```python
df_movies[df_movies['title'].isin(['Breaking Bad'])]
```

##### Research question 7
###### What show has the highest tmdb score

```python
Highest_imdb_score= df_movies.groupby(by = 'tmdb_score').max()[['title']].sort_values(by = ['title'], ascending = True).reset_index()
fig = plt.figure(figsize = (16,9))
plt.title("Movie Tmdb", size = 25)
axis = sns.barplot(data = Highest_imdb_score.iloc[:15], y = 'title', x = 'tmdb_score', linewidth = 2, edgecolor = 'green')
plt.xlabel('tmdb_score')
plt.ylabel('title')
plt.show
```

```python
df_movies[df_movies['title'].isin(['Power Rangers Beast Morphers'])]
```

##### From the above, Power Rangers Beast Morphers has the highest tmdb score
